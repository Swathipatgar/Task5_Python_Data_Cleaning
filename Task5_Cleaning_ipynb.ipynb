```python
import pandas as pd

df = pd.read_csv('/content/Titanic-Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.columns
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

```

```python
# Check number of duplicates first
print("Number of duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Verify duplicates removed
print("Number of rows after removing duplicates:", df.shape[0])
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
import numpy as np

df['Age_Group'] = np.where(df['Age'] < 18, 'Child', 'Adult')
```

```python
df.to_csv('/content/cleaned_data.csv', index=False)
```

```python
# Data Cleaning Notes

1. Loaded Titanic dataset using pandas.
2. Checked for missing values using df.isnull().sum().
3. Filled missing values in 'Age' with median.
4. Filled missing values in 'Embarked' with mode.
5. Removed duplicate rows to ensure data consistency.
6. Converted 'Age' column to integer for proper calculations.
7. Created new column 'Age_Group' to classify passengers as Child or Adult.
8. Saved cleaned dataset as 'cleaned_data.csv'._
```
